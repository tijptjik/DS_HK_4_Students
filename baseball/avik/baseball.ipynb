```python
df_HR_batting = baseballtraining[['HR_batting','salary']]
df_HR_batting.replace([np.inf, -np.inf], np.nan)
df_HR_batting.describe()
```

```python
for s in df_HR_batting[s]:
    if df_HR_batting[[s,'salary'] == 0:
        df_HR_batting.drop(df_HR_batting.index[s])
```

```python
df_HR_batting = df_HR_batting.drop(df_HR_batting[df_HR_batting.salary == 0].index)
df_HR_batting = df_HR_batting.drop(df_HR_batting[df_HR_batting.HR_batting == 0].index)
```
