```python
# Set some Pandas options
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)

# Default Plotting Size
mpl.rc("figure", figsize=(20, 8))

DATA_DIR = '../../../DS_HK_4/data/'
```

```python
# Load the baseball training and test datasets
bbtrain = pd.read_csv(DATA_DIR + 'baseball/baseball_training.csv')
bbtest = pd.read_csv(DATA_DIR + 'baseball/baseball_test.csv')

# Ensure all values numeric and replace any infinity or NaN values by zero 
bbtrain.convert_objects(convert_numeric=True)
bbtest.convert_objects(convert_numeric=True)

bbtrain = bbtrain.replace([np.inf, -np.inf, np.NaN], float(0))
bbtest = bbtest.replace([np.inf,-np.inf, np.NaN], float(0))
```

```python
# define functions
def SSE(pred, resp):
    return mean((pred - resp) ** 2)

def MSE(pred, resp):
    return metrics.mean_squared_error(pred, resp)

def fp_value(x,y):
    feature_selection.univariate_selection.f_regression(x, y)
```

```python
# Create sabermetric variables to use as regressors:
# 1."Slugging Percentage" +"On Base Percentage" = "On Base plus Slugging"
slg = lambda x: (x['H_batting']-x['2B']-x['3B']-x['HR_batting'] + 2*x['2B'] + 3*x['3B'] + 4*x['HR_batting'])/(x['AB']+1e-6)
bbtrain['SLG'] = bbtrain.apply(slg, axis=1)
bbtest['SLG'] = bbtest.apply(slg, axis=1)

obp = lambda x: (x['H_batting']+x['BB_batting']+x['HBP_batting'])/(x['AB']+x['BB_batting']+x['HBP_batting']+x['SF_batting']+1e-6) 
bbtrain['OBP'] = bbtrain.apply(obp, axis=1)
bbtest['OBP'] = bbtest.apply(obp, axis=1)

obs = lambda x: x['OBP'] + x['SLG']
bbtrain['OBS'] = bbtrain.apply(obs, axis=1)
bbtest['OBS'] = bbtest.apply(obs, axis=1)
```

```python
# 2. "Earned Run Average (ERA) metric, "Innings Pitched" and drop rows where "Innings Pitched"=0
ip = lambda x: x['IPouts']/3
bbtrain['IP'] = bbtrain.apply(ip, axis=1)
bbtrain = bbtrain[bbtrain['IP'] > 0]
bbtest['IP'] = bbtest.apply(ip, axis=1)
bbtest = bbtest[bbtest['IP'] > 0]
```

```python
era = lambda x: (9 * (x['ER']/x['IP']))
bbtrain['ERA'] = bbtrain.apply(era, axis=1)
bbtest['ERA'] = bbtest.apply(era, axis=1)
```

```python
# 3. "Fielding Independent Pitching (FIP)" metric and its constant
fipc = lambda x: (log(x['ERA'])-(((13*log(x['HR_pitching']))+(3*(log(x['BB_pitching'])+log(x['HBP_pitching'])))-(2*log(x['SO_pitching'])))/log(x['IP'])))
bbtrain['FIPC'] = bbtrain.apply(fipc, axis=1)
bbtest['FIPC'] = bbtest.apply(fipc, axis=1)

fip = lambda x: (x['FIPC'] + (((13*x['HR_pitching'])+(3*(x['BB_pitching']+x['HBP_pitching']))-(2*x['SO_pitching']))/x['IP']))
bbtrain['FIP'] = bbtrain.apply(fip, axis=1)
bbtest['FIP'] = bbtest.apply(fip, axis=1)

bbtrain = bbtrain.replace([np.inf, -np.inf, np.NaN], float(0))
bbtest = bbtest.replace([np.inf,-np.inf, np.NaN], float(0))
```

```python
# Using Statsmodel for linear regression model on sabermetric variables for
# batting, pitching, stolen bases & fielding (putouts, assists, double plays)
X = bbtrain[['OBP','FIP','SB_batting', 'PO', 'A', 'DP']]
Y = bbtrain[['log_salary']]

# Use Seaborn to plot scatterplots of hypothesized X variables & log salary
g = sns.PairGrid(X)
g.map(plt.scatter)
```

From scatterplots, it looks like the X-variables mostly appear independent except for 2 fielding statistics - Assists(A) & Double Plays(DP)

```python
# Using statsmodel to generate an OLS model 
est = sm.OLS(Y, X).fit()
est.summary()
```

The linear regression model generates a fairly high R-squared result, and all the selected X-variables are statistically significant in the training dataset.  This gives me confidence the model should work in predicting 2013 salaries on the test dataset

```python
# Statsmodel plot to compare the true relationship to OLS predictions with confidence intervals
prstd, iv_l, iv_u = wls_prediction_std(est)

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x, y, 'o', label="data")
ax.plot(x, y, 'b-', label="True")
ax.plot(x, est.fittedvalues, 'r--.', label="OLS")
ax.plot(x, iv_u, 'r--')
ax.plot(x, iv_l, 'r--')
ax.legend(loc='best');
```

```python
# Trying out ridge and polynomial models on training dataset
est1 = Ridge(normalize=True)
poly = PolynomialFeatures(3)

X1 = bbtrain[['OBP','FIP','SB_batting', 'PO', 'A', 'DP']].values
X1 = poly.fit_transform(X)
Y1 = bbtrain['log_salary'].values

# This runs a cross-validated routine (split-->fit--> predict--> score) 
scores = cross_val_score(est1, X1, Y1, cv=10)
np.mean(scores)
```

```python
# Using Statsmodel to predict salary in test dataset 
Xtest = bbtest[['OBP','FIP','SB_batting', 'PO', 'A', 'DP']]

ypred = est.predict(Xtest)
```
