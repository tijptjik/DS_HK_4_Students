```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('max_columns', 90)
pd.set_option('max_rows', 30)

CSV_DIR = '/vagrant/DS_HK_4/data/baseball/assignment/'
CODE_DIR = '/vagrant/DS_HK_4/code/student/ongoing/baseball/'
granularity = ['playerID', 'yearID', 'teamID']
```

```python
training_df = pd.read_csv(CSV_DIR + 'baseball_training.csv')
# training_df.set_index(granularity, inplace=True)
```

```python
training_df = training_df[training_df['salary'].notnull()]
```

```python
training_df = training_df[training_df['salary'] > 0]
```

```python
print training_df.salary.describe()
print training_df.salary.apply(np.log).describe()
plt.figure(figsize(20,8))
training_df.hist(column=['salary'], bins=50, alpha=0.3, normed=True, 
                   cumulative=True)
```

```python
# train_df_2012_subset = train_df_2012[train_df_2012.salary <= 500000] 
training_df['log_salary'] = training_df.salary.apply(np.log)

plt.figure(figsize(20,8))
training_df.hist(column=['log_salary'], bins=50, alpha=0.3, normed=True, 
                   cumulative=True)
```

```python
training_df['experience'] = training_df.yearID - training_df.debut.apply(lambda x: float(x[:4]))
```

```python
colors = matplotlib.rcParams['axes.color_cycle']
for i, year in enumerate([2008, 2009, 2010, 2011, 2012]):
    plt.figure(figsize(20,8))
    x = training_df[training_df.yearID == year].experience
    y = training_df[training_df.yearID == year].salary / (10 ** 6)
    color = colors[i]
    plt.scatter(x, y, color=color, alpha=.2)
    plt.title('Year %d' % year)
    plt.xlabel('year of experience')
    plt.yscale('log')
    plt.ylabel('salary (million) on log')
    plt.show()
```

From 2008 to 2012, players with experience less than 5 years were more likely to get lower salaries (close to bottom), while senior players generally earned more.

```python
plt.figure(figsize(20, 8))
plt.scatter(training_df.yearID, training_df.log_salary, color='g',
            alpha=0.3)
plt.show()
```

The relation between yearID and salary is quite clear (because of inflation?).

```python
# train_df_2012[['experience','log_salary']].info()
from sklearn import linear_model, cross_validation
from sklearn.preprocessing import PolynomialFeatures

def SSE(pred, resp):
    return np.mean((pred - resp) ** 2)
```

```python
# a (over) simple model considering only 1 variable

regr = linear_model.LinearRegression()
X = training_df[['experience']].values
y = training_df.log_salary.values
regr.fit(X, y)

print "SSE: %.4f" % (SSE(regr.predict(X), y))
print "R^2: %.4f" % (regr.score(X, y))
```

```python
X = training_df[['yearID', 'experience']]
X['exp_sq'] = X['experience'] ** 2
X = X.values
y = training_df.log_salary.values

ridge = linear_model.Ridge()

train_sse, train_r_sq, test_sse, test_r_sq = [], [], [], []
for i in range(20):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, 
                                        test_size=.3)
    ridge.fit(X_train, y_train)
    train_sse.append(SSE(ridge.predict(X_train), y_train))
    train_r_sq.append(ridge.score(X_train, y_train))
    test_sse.append(SSE(ridge.predict(X_test), y_test))
    test_r_sq.append(ridge.score(X_test, y_test))

# print train_r_sq
print "---- train score ----"
print "MSE: %.4f" % (np.mean(train_sse))
print "R^2: %.4f" % (np.mean(train_r_sq))

print "---- test score ----"
print "MSE: %.4f" % (np.mean(test_sse))
print "R^2: %.4f" % (np.mean(test_r_sq))
```
