```python
df['log_salary'] = log(df.salary)
df.columns
```

```python
def hasPrevSalary(r):
    try:
        prev = df.xs((r['playerID'], int((r['yearID'] - 1))))['log_salary'].values
    except:
        return df.xs((r['playerID'], int((r['yearID']))))['log_salary'].values[0]  #if no prev year, just use this year
    if len(prev) == 0:
        return df.xs((r['playerID'], int((r['yearID']))))['log_salary'].values[0]  #if no prev year, just use this year
    else:
        return prev[0]
    
```

```python
df = df.set_index(['playerID','yearID','teamID'])
df.head()
```

```python
prev_year_log_salary = df.reset_index()[['playerID','yearID','teamID','log_salary']].apply(hasPrevSalary, axis=1)
```

Note that in the above, the reset_index() did NOT change the multi-index of df since it didn't reassign nor use inplace = True. It just created another view that has the index reset and brought back the 4 necessary cols

```python
type(prev_year_log_salary)
```

```python
df.reset_index(inplace=True)
```

```python
# Add new column
df['prev_year_log_salary'] = prev_year_log_salary
```

```python
df.info()
```

```python
from sklearn import linear_model, metrics
```

```python
train_X = df[['yearID','prev_year_log_salary']][(df.yearID > 2010)&(df.yearID < 2012)].values
train_y = df['log_salary'][(df.yearID > 2010)&(df.yearID < 2012)].values
```

```python
lm = linear_model.Ridge()
lm.fit(train_X, train_y)
```

```python
print 'R-Squared:',lm.score(train_X, train_y)

print 'MSE:',metrics.mean_squared_error(lm.predict(train_X), train_y)
```
