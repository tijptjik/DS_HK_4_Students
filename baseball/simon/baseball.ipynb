```python
# Try to evaluate which features have significant contribution and narrow down the feature space

regressors=train_v2[overlap_features_2].values
target=train_v2['salary'].values

#print regressors[:5]
#print target[:5]
fp_value = feature_selection.univariate_selection.f_regression(regressors, target)
p_value = zip(overlap_features_2,fp_value[1])

sorted_p_value = sorted(p_value,key=lambda x: x[1])
print sorted_p_value
best = [x[0] for x in sorted_p_value if x[1]<0.01]
print len(best)
print sorted(best)
```

```python
# Model 1: Simple Linear model on features_set

regressors=train_v2[best].values
target=train_v2['salary'].values

regr = linear_model.LinearRegression()
regr.fit(regressors,target)
pred=regr.predict(regressors)

# Plot the predicted value Vs actual
salary=train_v2['salary'].values
x=[x for x in range(22116)]
plt.plot(x,target,'r.',x,pred,'b-')
plt.show()

print "R-Squared: %.4f:" %(regr.score(regressors,target))
print 'MSE: %.4f'%(mse(pred,salary))

# R-Squared: 0.7894:
# MSE: 1873907673825.7253
```

```python
# Model 2: Ridge model on overlap feature set

regressors=train_v2[best].values
target=train_v2['salary'].values

r1=linear_model.Ridge(alpha=1,normalize=True)
r1.fit(regressors,target)
pred=r1.predict(regressors)

# Plot the predicted value Vs actual
salary=train_v2['salary'].values
x=[x for x in range(22116)]
plt.plot(x,target,'r.',x,pred,'b-')
plt.show()

print "R-Squared: %.4f:" %(r1.score(regressors,target))
print 'MSE: %.4f'%(metrics.mean_squared_error(pred, salary))

# R-Squared: 0.6194:
# MSE: 3386369126065.4004
```
