# Scoring Baseball Submissions

```python
import os, sys

# Getting submissions
ls = os.listdir('.')
submissions = [d for d in ls if os.path.isdir(d) and not d.startswith('.')]
submissions
```

## Load the Submissions

```python
import pandas as pd

results = {}

for submission in submissions:
    path = submission + '/submission.csv'
    if os.path.isfile(path):
        results[submission] = pd.read_csv(path)
```

## Load the Results

```python
actual = pd.read_csv('baseball_results.csv',  header=None, names=['playerID','yearID','teamID','salary'])    
```

```python
actual = actual.dropna()
```

```python
actual = actual.set_index('playerID').drop(['yearID','teamID'], axis=1)
```

```python
len(actual.index) == len(set(actual.index))
```

```python
actual.head()
```

## Clean Submissions

```python
clean_submissions = {}
for name, prediction in results.iteritems():
    pred = prediction.columns[-1] 
    df = prediction[['playerID', pred]]
    df.columns = ['playerID', 'predicted']
    clean_submissions[name] = df
```

```python
clean_submissions
```

## Scoring Metric

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score as R2
```

## Score the Submissions

```python
scores = {}
for name, submission in clean_submissions.iteritems():
    df = pd.merge(actual, submission, how='inner', left_index=True, right_on='playerID')
    y_true = df['salary'] 
    y_pred = df['predicted']
#     print name, MSE(y_pred, y_true) / 10000000 , R2(y_true, y_pred)
    scores[name] = MSE(y_pred, y_true), R2(y_true, y_pred)
```

```python
df = pd.DataFrame(scores).T
df.columns = ['MSE','R2']
df
```
